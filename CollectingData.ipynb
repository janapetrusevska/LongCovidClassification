```python
from mpl_toolkits.mplot3d import Axes3D

import seaborn as sns

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data20 = pd.read_csv('/content/drive/MyDrive/finki/StandardenProekt VNP/datasets/lc2020.csv', encoding='latin1')
```

```python
longCovidTweets20 = data20[data20['Tweets'].str.contains('#LongCovid', case=False, na=False)]
```

```python
data20.info()
```

```python
longCovidTweets20.info()
```

We only take 60 952 tweets out of 158 280 tweets (38.5%) from 2020.

```python
longCovidTweets20['Date'] = pd.to_datetime(longCovidTweets20['Date'])

monthly_distribution = longCovidTweets20['Date'].dt.month.value_counts().sort_index()

plt.figure(figsize=(20, 6))
sns.lineplot(x=monthly_distribution.index, y=monthly_distribution.values, color='green')

plt.xlabel('Month')
plt.ylabel('Tweet Count')
plt.title('Monthly Distribution of 2020 Tweets', fontsize=15)

plt.tight_layout()
plt.show()
```

We can see that in **october** 2020 there are the most tweets with #LongCovid (over 14 000)
